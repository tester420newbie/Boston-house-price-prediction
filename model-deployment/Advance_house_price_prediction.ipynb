## independent and dependent feature

```python
x = data.iloc[:,:-1]
y = data["MEDV"]
```

```python
print(x)
```

Train test split

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=42,test_size=0.3 )
```

#standardizing data

```python
from sklearn.preprocessing import FunctionTransformer
func = FunctionTransformer(func=np.log1p)
x_train = func.fit_transform(x_train)
x_test = func.transform(x_test)
```

```python
for feature in features :
    sns.distplot(data[feature])
    plt.title(feature)
    plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

## Model Training 

```python
from sklearn.linear_model import LinearRegression
regression = LinearRegression()
```

```python
regression.fit(x_train,y_train)
```

Print the coefficients

```python
print(regression.coef_)
```

```python
print(regression.intercept_)
```

On which parameter the model has been train

```python
print(regression.get_params())
```

Predictions

```python
reg_pre = regression.predict(x_test)
```

```python
print(reg_pre)
```

```python
sns.scatterplot(x=reg_pre,y=y_test,markers="o")
plt.title("Prediction and Actual value ")
plt.grid()
plt.show()
```

```python
residuals = y_test-reg_pre
```

```python
print(residuals)
```

```python
sns.displot(residuals,kind="kde")
plt.title("Residual plot")
plt.show()
```

```python
sns.scatterplot(x=reg_pre,y=residuals)
plt.title("Prediction and Residual PLot")
plt.show()
```

```python
from sklearn.metrics import mean_squared_error , mean_absolute_error
```

```python
print(mean_absolute_error(y_test,reg_pre))
```

```python
print(mean_squared_error(y_test,reg_pre))
```

```python
print(regression.score(x_train,y_train)*100)
print(regression.score(x_test,y_test)*100)
```
